```python
import os
import csv
import random
import pandas as pd
import gensim
import logging
 
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

### DataSet Used - IMDB review dataset
- Link - https://www.kaggle.com/utathya/imdb-review-dataset

```python
file_name = 'imdb_master.csv'

file_path = os.path.join(os.getcwd(), file_name)
print(file_path)
```

```python
df = pandas.read_csv(file_path, encoding='latin-1')
```

```python
def process_text():
    for i, line in enumerate(df['review']):
        if (i % 10000 == 0):
            logging.info("read {0} reviews".format(i))
        yield gensim.utils.simple_preprocess(line)
```

```python
processed_text = list(process_text())
```

```python
 model = gensim.models.Word2Vec(processed_text, size=100, window=5, 
                                min_count=2, workers=8, seed=10, iter=3, negative=10)
    
# size=300, window=5, dm=0, dbow_words=1, hs=0, negative=10, alpha=0.05, min_count=20, workers=cores, sample=1e-5, seed=0, iter=10    
# model.train(documents, total_examples=len(documents), epochs=10)
```

```python
model.save(os.path.join(os.getcwd(), 'trained_model'))
```

```python
print(model.similarity.__doc__)
```

```python
keys = set(model.vocab.keys())
```

# Task - 1 - Similarity between 2 words

```python
def similarity(word1, word2):
    if word2 in keys and word2 in keys:
        return model.similarity(word1, word2)
    else:
        print("Either or both of the words are not present in model vocabulary")
```

```python
words_list = [random.sample(keys, 2) for x in range(20)]
for item in words_list:
    word1, word2 = item[0], item[1]
    score = similarity(word1, word2)
    print("Similarity between {:<15} & {:<15} is: {}".format(word1, word2, score))
```
